# 1 A simple plot

```python
from matplotlib import pyplot as plt
import numpy as np
```

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 4, 8, 12, 16, 20] #x values
y_1 = [1, 2, 3, 4, 5, 6] #y value (set 1)
y_2 = [5, 6, 7, 8, 9, 10] #y value (set 2)

# Let's start plotting
plt.plot(x, y_1, label='Y values', #plots x against y_1, labelling line to y values
         color='purple', linestyle='dashed') #set colour to red and format as dash
plt.xlabel('x-values') #label x axis
plt.ylabel('y-values') #label y axis
plt.title('X vs Y') #label graph
plt.grid(alpha=.25) #adding grid lines, alpha is transparency
plt.legend(loc='upper left') #adding legend to plot with location at upper left
plt.show
```

### Things to note

| Long form | Abbreviation |
|:---------:|:------------:|
|   color   |       c      |
| linestyle |      ls      |
| linewidth |      lw      |

```python
plt.plot(x, y_1, label='Y values', #plots x against y_1, labelling line to y values
         c='blue', ls='dashed', lw=2) #set colour to red and format as dash
plt.show() #good to show
```

```python
plt.plot(x, y_1, y_2) #two y sets so two lines produced, default formatting
plt.show()
```

- Jupyter is an interactive environment so can output even if you omit plt.show(). However, it is good practice to include this line
- plotting functions usually have default values for the styling parameters
- can split the arguments into separate lines to improve readability
- order of keyword arguments does not matter (linewidth, linestyle)

## 1.2 Adding another plot

```python
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue') #add another line of code to plot data of y_2 and format it
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

## 1.3 Yet another plot but with error bars

```python
y_3 = [7, 8, 9, 10, 11, 12]
x_error = 0.5 #uncertainty about x
y_error = 1 #uncertainty about y

plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
```

can also provide a list of errors so that each will have a different length

# 2 Better with NumPy

```python
import numpy as np
np_x = np.array(x) #converting to numpy arrays
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

## 2.1 Adding mathematical functions

NumPy arrays allow us to generate data-related mathematical functions easily

```python
x = np.array([0, 2, 4, 6, 7, 8])

x3 = x**3   # array can do maths
cos_x = np.cos(x)

plt.plot(x, x3, label='x^3',
         color='yellow', linestyle='dashed', )
plt.plot(x, cos_x, label='cos(x)',
         color='blue')
plt.legend(loc="upper right")  
plt.show()
```

plot does not look good because lies between +/-1 but x^3 has no such bounds. fix by adding another y-axis that shares the same x-axis.

## We need another axis!

```python
x = np.array([0, 2, 4, 6, 7, 8])

x3 = x**3   
cos_x = np.cos(x)

plt.plot(x, x3, label='x^3', 
         color='yellow', linestyle='dashed', ) #first y-axis

plt.twinx() #adding another axis

plt.plot(x, cos_x, label='cos(x)',
         color='blue') #second y-axis
plt.legend(loc="upper right")  
plt.show()
```

### Things to note

```python
x = np.linspace(0, 8, num=100) #generates 100 evenly spaced x-values between 0 and 8 so more points and a smoother curve
x3 = x**3   
cos_x = np.cos(x)

plt.plot(x, x3, label='x^3', 
         color='yellow', linestyle='dashed', ) #first y-axis
plt.legend(loc="upper left") #first legend

plt.twinx() #adding another axis

plt.plot(x, cos_x, label='cos(x)',
         color='blue') #second y-axis
plt.legend(loc="upper right")  #second legend
plt.show()
```

# 3 Saving to disc

to use your plot in a report or presentation, you must first save it to disk and export plots into many formats (PDF, JPEG, PNG, BMP, etc.) and at different resolutions.

- use the function savefig() for this.
- specify the format with the extension (e.g., filename.pdf) of the file name,
- specify resolution by use dots-per-inch (dpi).
- When you run this code, file saved in the same directory (folder) as the one in your notebook lives in. If you want it saved elsewhere, specify the path in more detail

```python
x = np.linspace(0, 8, num=100) #generates 100 evenly spaced x-values between 0 and 8 so more points and a smoother curve
x3 = x**3   
cos_x = np.cos(x)

plt.plot(x, x3, label='x^3', 
         color='yellow', linestyle='dashed', ) #first y-axis
plt.legend(loc="upper left") #first legend

plt.twinx() #adding another axis

plt.plot(x, cos_x, label='cos(x)',
         color='blue') #second y-axis
plt.legend(loc="upper right")  #second legend
plt.savefig('C:/Users/awyet/OneDrive/Documents/NUS/Y1S2/SP2273/learning-portfolio-Yeting2004/plotting/my_first_plot.png', dpi=150)
plt.show() #only show after saving, if not python will save blank figure
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt', #load data from text file
                   skiprows=24) #skips first 24 rows as not relavant
date = data[:, 0] #extracting data from first column
anomaly = data[:, -1] #extracting data from last column

plt.plot(date, anomaly, alpha=.5) 
plt.ylim([-8, 8]) #set y-axis limits
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') #adding horizontal line at y=0
```

## 4.2 Styles

Matplotlib offers some standard style [templates](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html)

```python
plt.style.use('fivethirtyeight') #must be at the front

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
