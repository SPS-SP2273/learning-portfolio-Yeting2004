## Exercise 1

```python
from matplotlib import pyplot as plt
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5, #ecolor is specifying error bar error
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

## Exercise 2

```python
import numpy as np
x = np.linspace(0, 50, num=200)
y = np.exp(-x/10) *np.sin(x)
plt.plot(x, y)
plt.title(r'$y = e^{-x/10} sin(x)$')
plt.ylim(-1,1)
plt.xlabel("x", color="darkorange", fontsize = 15)
plt.ylabel("y", color="darkorange", fontsize = 15, rotation=0)
plt.grid(alpha = 0.25)
plt.show()
```
